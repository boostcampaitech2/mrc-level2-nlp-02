### kobert로 임베딩을 만들어보자

```python
import torch
from transformers import BertModel
from kobert_tokenizer import KoBERTTokenizer

tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')
model = BertModel.from_pretrained('skt/kobert-base-v1')
```

```python
# question을 string으로 구성된 리스트로 넣기
good_ppg_q = good_ppg['question'].to_list()
```

```python
tokenized_good_ppg_q = tokenizer.batch_encode_plus(good_ppg_q, padding=True)
```

```python
good_ppg_out = model(input_ids = torch.tensor(tokenized_good_ppg_q['input_ids']),
            attention_mask = torch.tensor(tokenized_good_ppg_q['attention_mask']))
```

```python
good_ppg_vec = good_ppg_out.pooler_output
```

```python
# raw data (reference) 가져오기
good_ref = []
bad_ref = []

for i in good:
    good_ref.append(raw_dataset[i]['question'])

for k in bad:
    bad_ref.append(raw_dataset[k]['question'])
```

```python
tokenized_good_ref = tokenizer.batch_encode_plus(good_ref, padding=True)
good_ref_out = model(input_ids = torch.tensor(tokenized_good_ref['input_ids']),
            attention_mask = torch.tensor(tokenized_good_ref['attention_mask']))
```

```python
good_ref_vec = good_ref_out.pooler_output
```

```python
good_ppg_vec.size()
```

```python
# from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))
```

```python
for i in range(10):
    print(f'{i}th sentence cosine similarity: ', cos_sim(good_ppg_vec[i].detach().numpy(), good_ref_vec[i].detach().numpy()))
```

```python
check_index = 6
print('reference sentence:  ', good_ref[check_index])
print('papago rtt sentence: ', good_ppg_q[check_index])
```
